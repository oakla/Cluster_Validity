```python
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN

import Toolbox.clustering_tools as clstr_tools
import Toolbox.plotting_functions as plt_tools
import numpy as np

import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import scale
X, y = load_iris(return_X_y=True)
X_scaled = scale(X)
n_samples, n_features = X.shape

min_pts = 8
eps_list = np.linspace(0.25,0.75,11)
metrics=['silhouette', 's_dbw']

metric_scores_dict, dbs_list = clstr_tools.benchmark_dbscan(X_scaled, 
                                                            min_pts=min_pts, 
                                                            eps_list=eps_list, 
                                                            print_rslt=False,
                                                            validity_metrics=metrics)

```

```python
plt.figure(figsize=(9,6))
for metric in metrics:
    scores = metric_scores_dict[metric]
    plt.plot(eps_list, scores, label=metric)
plt.legend()
plt.title("Scores of DBSCAN Clusterings. Min_points = 8")
plt.axis([0.2,0.8,0,0.8])
plt.xticks(eps_list)
plt.xlabel("Epsilon")
plt.ylabel("Score")
plt.grid()
plt.show()
```

```python
for i in range(len(dbs_list)):
    params = dbs_list[i].get_params()
    plt_tools.visualize_clusters(X_scaled,dbs_list[i].labels_)
    plt.title("Minium Points= {0}, Epsilon={1}".format(params['min_samples'], params['eps']) +
                "\nSilhouette: {0},    S_Dbw: {1}".format(round(metric_scores_dict['silhouette'][i],2),
                                                     round(metric_scores_dict['s_dbw'][i],2)))
                
```
