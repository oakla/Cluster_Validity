Here we simply run HDBSCAN with a series of different min_cluster_size arguments. This is the main parameter of HDBSCAN.

```python
import hdbscan

import Toolbox.plotting_functions as plt_tools
import Toolbox.clustering_tools as clstr_tools

from sklearn.datasets import load_iris
from sklearn.preprocessing import scale
```

```python
X, y = load_iris(return_X_y=True)
X_scaled = scale(X)
n_samples, n_features = X.shape
```

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=2)
est_labels = clusterer.fit_predict(X_scaled)
```

```python
plt_tools.visualize_clusters(X, est_labels , dims=2)
```

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=5)
est_labels = clusterer.fit_predict(X_scaled)
plt_tools.visualize_clusters(X, est_labels , dims=2)
```
