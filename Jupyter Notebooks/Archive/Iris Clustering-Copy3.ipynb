## Visualizing The Data
### PCA
Here we use PCA to reduce the dimensions of the Iris data to something that we can visualize more easily.
Below is code to create 3D and 2D plots

```python
def visualize_labelled_data(data=None, X=None, y=None, dims=2, names=None):
    if data != None:
        X = scale(data.data)
        y = data.target
        names = data.target_names
    pca = decomposition.PCA(n_components=dims)
    pca.fit(X)
    X = pca.transform(X)
    
    fig = plt.figure(figsize=(8, 8))
    
    if dims == 2:
        ax = fig.add_subplot(1,1,1)
    elif dims == 3:
        ax =  Axes3D(fig, rect=[0, 0, .95, 1])
 
    plt.cla()
    
    ax.scatter(X[:,0], X[:,1], c=y.astype(float), cmap='Set1')
    ax.set_xlabel('Principal Component 1', fontsize=15)
    ax.set_ylabel('Principal Component 2', fontsize=15)
    ax.set_title('2 Component PCA of Ground Truth', fontsize=20)

    ax.grid()
```

```python
# 2D Plot
wine = load_wine()

visualize_labelled_data(data=wine)
```
